```python
# ============================
# Self-Organizing Maps (SOM)
# ============================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_circles, make_swiss_roll

# -----------------------------
# 1) Implementación básica SOM
# -----------------------------
class SOM:
    """
    Self-Organizing Map (SOM) rectangular 2D, entrenado con descenso competitivo.
    - grid: (m, n) neuronas en una rejilla 2D.
    - dim: dimensión de los vectores de entrada.
    - lr0: tasa de aprendizaje inicial.
    - sigma0: radio inicial de vecindad (en celdas de la rejilla).
    Entrena por épocas barajando los patrones.
    """
    def __init__(self, m, n, dim, lr0=0.5, sigma0=None, seed=42):
        self.m, self.n, self.dim = m, n, dim
        self.rng = np.random.default_rng(seed)
        # Pesos/prototipos (m*n, dim)
        self.W = self.rng.normal(loc=0.0, scale=1.0, size=(m*n, dim))
        # Coordenadas de las neuronas en la rejilla
        xs, ys = np.meshgrid(np.arange(m), np.arange(n), indexing='ij')
        self.coords = np.stack([xs.ravel(), ys.ravel()], axis=1)
        # Hiperparámetros
        self.lr0 = lr0
        self.sigma0 = sigma0 if sigma0 is not None else max(m, n) / 2.0

    @staticmethod
    def _euclidean(a, b):
        # a: (N, d), b: (M, d) -> distancias (N, M)
        # Usamos broadcasting para eficiencia
        a2 = np.sum(a*a, axis=1, keepdims=True)
        b2 = np.sum(b*b, axis=1, keepdims=True)
        return np.sqrt(a2 + b2.T - 2*np.dot(a, b.T) + 1e-12)

    def _bmu_index(self, x):
        # Índice del "Best Matching Unit" para cada x
        d = self._euclidean(x, self.W)  # (N, m*n)
        return np.argmin(d, axis=1)     # (N,)

    def _decays(self, t, T):
        # Decaimiento exponencial de lr y sigma a lo largo de T pasos
        lr = self.lr0 * np.exp(-t / (T/np.log(2)))
        sigma = self.sigma0 * np.exp(-t / (T/np.log(2)))
        return lr, max(sigma, 1e-6)

    def fit(self, X, epochs=20, verbose=True):
        """
        Entrenamiento online (muestra por muestra) con vecindad gaussiana.
        """
        N = X.shape[0]
        total_steps = epochs * N
        step = 0
        if verbose:
            print(f"[SOM] Entrenando: grid={self.m}x{self.n}, dim={self.dim}, epochs={epochs}, muestras={N}")
        for ep in range(epochs):
            idx = self.rng.permutation(N)
            for i in idx:
                x = X[i:i+1]  # (1, dim)
                # 1) Encontrar BMU
                bmu = self._bmu_index(x)[0]
                # 2) Decaimiento de hiperparámetros
                lr, sigma = self._decays(step, total_steps)
                # 3) Kernel gaussiano de vecindad en la rejilla
                bmu_coord = self.coords[bmu]  # (2,)
                # Distancia euclidiana en la rejilla
                grid_dist2 = np.sum((self.coords - bmu_coord)**2, axis=1)
                h = np.exp(-grid_dist2 / (2*sigma*sigma))  # (m*n,)
                # 4) Actualización: W_j <- W_j + lr * h_j * (x - W_j)
                # Expandimos dimensiones para broadcasting
                h = h[:, None]                       # (m*n, 1)
                self.W += lr * h * (x - self.W)     # (m*n, dim)
                step += 1
            if verbose:
                print(f"  - Epoch {ep+1}/{epochs}: lr={lr:.4f}, sigma={sigma:.4f}")

    # --------- Utilidades ---------
    def map_vectors(self, X):
        """Retorna coordenadas (fila, col) del BMU de cada vector X."""
        bmu = self._bmu_index(X)
        rows = bmu // self.n
        cols = bmu % self.n
        return np.stack([rows, cols], axis=1)

    def u_matrix(self):
        """
        U-Matrix: para cada neurona, promedio de distancias a sus vecinos 4-conectados.
        Mide "fronteras" (valores altos ~ límites entre clusters).
        """
        U = np.zeros((self.m, self.n))
        Wgrid = self.W.reshape(self.m, self.n, self.dim)
        for i in range(self.m):
            for j in range(self.n):
                neigh = []
                for di, dj in [(1,0),(-1,0),(0,1),(0,-1)]:
                    ii, jj = i+di, j+dj
                    if 0 <= ii < self.m and 0 <= jj < self.n:
                        neigh.append(np.linalg.norm(Wgrid[i,j]-Wgrid[ii,jj]))
                U[i,j] = np.mean(neigh) if len(neigh)>0 else 0.0
        return U

    def prototypes(self):
        """Devuelve los prototipos en forma (m, n, dim)."""
        return self.W.reshape(self.m, self.n, self.dim)

# --------------------------------------
# 2) Helpers de visualización explicados
# --------------------------------------
def plot_u_matrix(som, title="U-Matrix"):
    U = som.u_matrix()
    plt.figure(figsize=(5,5))
    plt.imshow(U.T, origin='lower', interpolation='nearest')
    plt.colorbar(label='Distancia media a vecinos')
    plt.title(title)
    plt.xlabel("Fila (i)")
    plt.ylabel("Columna (j)")
    plt.show()

def plot_hits(som, X, y=None, title="Hits por neurona"):
    """
    Cuenta cuántos vectores caen en cada neurona (BMU).
    Si se pasa y (clases), muestra la clase mayoritaria por neurona.
    """
    hits = np.zeros((som.m, som.n), dtype=int)
    labels_map = [[[] for _ in range(som.n)] for _ in range(som.m)]
    bmu_rc = som.map_vectors(X)
    for k,(r,c) in enumerate(bmu_rc):
        hits[r,c] += 1
        if y is not None:
            labels_map[r][c].append(y[k])

    plt.figure(figsize=(6,5))
    plt.imshow(hits.T, origin='lower', interpolation='nearest')
    plt.colorbar(label='# Asignaciones')
    plt.title(title)
    plt.xlabel("Fila (i)")
    plt.ylabel("Columna (j)")
    plt.show()

    if y is not None:
        # Mapa de clase mayoritaria (simple inspección)
        majority = np.full((som.m, som.n), fill_value=-1, dtype=int)
        for i in range(som.m):
            for j in range(som.n):
                if len(labels_map[i][j])>0:
                    vals, counts = np.unique(labels_map[i][j], return_counts=True)
                    majority[i,j] = vals[np.argmax(counts)]
        plt.figure(figsize=(6,5))
        plt.imshow(majority.T, origin='lower', interpolation='nearest')
        plt.title("Clase mayoritaria por neurona (–1 = vacía)")
        plt.xlabel("Fila (i)")
        plt.ylabel("Columna (j)")
        plt.colorbar()
        plt.show()

def plot_prototypes_2d(som, X=None, title="Prototipos SOM (2D)"):
    """
    Si los prototipos son 2D, grafica su malla y opcionalmente datos X.
    """
    W = som.prototypes()  # (m, n, dim)
    if W.shape[2] != 2:
        print("Los prototipos no son 2D; omitiendo esta figura.")
        return
    plt.figure(figsize=(6,6))
    if X is not None and X.shape[1]==2:
        plt.scatter(X[:,0], X[:,1], s=10, alpha=0.3, label='Datos')
    # Conectamos la rejilla de prototipos
    for i in range(som.m):
        plt.plot(W[i,:,0], W[i,:,1], '-')
    for j in range(som.n):
        plt.plot(W[:,j,0], W[:,j,1], '-')
    plt.title(title)
    plt.axis('equal')
    plt.legend()
    plt.show()

def auto_grid(n_samples):
    """
    Heurística simple para elegir tamaño de rejilla: ~ 5*sqrt(N) neuronas.
    Distribuye en m x n lo más cuadrado posible.
    """
    k = int(5*np.sqrt(n_samples))
    k = max(k, 16)  # mínimo
    m = int(np.sqrt(k))
    n = int(np.ceil(k/m))
    return m, n

# ------------------------------------
# 3) Datasets sintéticos y demostrador
# ------------------------------------
def demo_blobs():
    print("\n=== DEMO 1: Clustering en blobs 2D ===")
    X, y = make_blobs(n_samples=1500, centers=4, cluster_std=1.2, random_state=7)
    m, n = auto_grid(len(X))
    som = SOM(m, n, dim=2, lr0=0.4, sigma0=max(m,n)/2)
    som.fit(X, epochs=20)

    print("• U-Matrix: resalta fronteras entre clústeres.")
    plot_u_matrix(som, "U-Matrix (Blobs)")
    print("• Hits: cuántos puntos cayeron en cada neurona (densidad de datos).")
    plot_hits(som, X, y, "Hits y clases (Blobs)")
    print("• Prototipos: cómo la rejilla SOM se amolda al espacio de datos.")
    plot_prototypes_2d(som, X, "Prototipos en 2D (Blobs)")

def demo_circles():
    print("\n=== DEMO 2: Datos no lineales (círculos concéntricos) ===")
    X, y = make_circles(n_samples=1800, factor=0.45, noise=0.06, random_state=3)
    m, n = auto_grid(len(X))
    som = SOM(m, n, dim=2, lr0=0.5, sigma0=max(m,n)/2)
    som.fit(X, epochs=25)

    print("• U-Matrix: ayuda a ver la separación anular (no lineal).")
    plot_u_matrix(som, "U-Matrix (Círculos)")
    plot_hits(som, X, y, "Hits y clases (Círculos)")
    plot_prototypes_2d(som, X, "Prototipos en 2D (Círculos)")

def demo_swiss_roll_dimred():
    print("\n=== DEMO 3: Reducción de dimensionalidad (Swiss roll 3D → mapa 2D) ===")
    X, t = make_swiss_roll(n_samples=2000, noise=0.05, random_state=10)
    # SOM 2D con datos 3D aprende la “topología” y la despliega en la rejilla.
    m, n = auto_grid(len(X))
    som = SOM(m, n, dim=3, lr0=0.4, sigma0=max(m,n)/2)
    som.fit(X, epochs=25)

    print("• U-Matrix en el espacio de prototipos 3D (proyectada como U-Matrix 2D).")
    plot_u_matrix(som, "U-Matrix (Swiss roll 3D)")
    # Proyección: coloreamos neuronas por el 't' medio de sus asignaciones para ver el desenrollado
    bmu = som.map_vectors(X)
    color_grid = np.full((som.m, som.n), np.nan)
    accum = [[[] for _ in range(som.n)] for _ in range(som.m)]
    for i,(r,c) in enumerate(bmu):
        accum[r][c].append(t[i])
    for i in range(som.m):
        for j in range(som.n):
            if len(accum[i][j])>0:
                color_grid[i,j] = np.mean(accum[i][j])

    plt.figure(figsize=(6,5))
    im = plt.imshow(color_grid.T, origin='lower', interpolation='nearest')
    plt.title("‘Desenrollado’ por color (promedio de t por neurona)")
    plt.xlabel("Fila (i)"); plt.ylabel("Columna (j)")
    plt.colorbar(im, label='t promedio (color de la swiss roll)')
    plt.show()

def demo_color_quantization():
    print("\n=== DEMO 4: Cuantización de color (imagen sintética) ===")
    # Creamos una “imagen” sintética con regiones de color suaves (256x256)
    H, W = 180, 240
    x = np.linspace(0,1,W)
    y = np.linspace(0,1,H)
    xx, yy = np.meshgrid(x, y)
    # Gradientes que generan colores
    img = np.stack([
        0.6*xx + 0.4*yy,     # R
        0.3*(1-xx) + 0.7*yy, # G
        0.5*np.sin(2*np.pi*xx)*0.25 + 0.5*yy + 0.25 # B
    ], axis=2)
    img = np.clip(img, 0, 1)

    # Aplanamos a una nube de puntos RGB
    X = img.reshape(-1, 3)

    # Creamos un SOM pequeño para “paleta” de colores
    m, n = 10, 10   # 100 colores prototipo
    som = SOM(m, n, dim=3, lr0=0.5, sigma0=max(m,n)/2)
    som.fit(X, epochs=8)

    # Asignamos cada píxel a su prototipo más cercano
    W = som.W  # (100, 3)
    d = SOM._euclidean(X, W)     # (Npix, 100)
    idx = np.argmin(d, axis=1)   # (Npix,)

    print(f"Shape of W: {W.shape}, Dtype of W: {W.dtype}")
    print(f"Shape of d: {d.shape}, Dtype of d: {d.dtype}")
    print(f"Shape of idx: {idx.shape}, Dtype of idx: {idx.dtype}")

    # Explicitly cast idx to int to avoid TypeError
    # Attempting indexing. The error occurs here.
    try:
        quant = W[idx.astype(int)].reshape(H, W, 3)
        quant = np.clip(quant, 0, 1)

        # Visualización
        plt.figure(figsize=(10,4))
        plt.subplot(1,2,1); plt.imshow(img); plt.title("Imagen sintética original"); plt.axis('off')
        plt.subplot(1,2,2); plt.imshow(quant); plt.title("Cuantización SOM (100 colores)"); plt.axis('off')
        plt.show()

        print("• La U-Matrix en RGB muestra continuidad/discontinuidades en la paleta.")
        plot_u_matrix(som, "U-Matrix (Paleta de color SOM)")

    except TypeError as e:
        print(f"TypeError during quantization: {e}")
        print("W shape:", W.shape)
        print("W dtype:", W.dtype)
        print("idx shape:", idx.shape)
        print("idx dtype:", idx.dtype)
        # Optionally, print a snippet of idx to see its values
        print("idx[:10]:", idx[:10])


# ------------------------
# 4) Ejecutar todas las demos
# ------------------------
print("====================================================")
print("SOM demos: explicación breve")
print("Un SOM es una red competitiva que mapea datos de alta/igual dimensión")
print("a una rejilla 2D preservando topología local. Útil para clustering,")
print("reducción de dimensionalidad, visualización y cuantización de vectores.")
print("Hiperparámetros clave: tasa de aprendizaje (lr) y radio de vecindad (sigma).")
print("====================================================")

demo_blobs()
demo_circles()
demo_swiss_roll_dimred()
demo_color_quantization()

print("\nFin de las demostraciones. Sugerencias:")
print("- Ajusta epochs, lr0 y sigma0 para observar convergencia más/menos suave.")
print("- Cambia el tamaño de rejilla (m, n) para mayor/menor resolución del mapa.")
print("- Prueba con tus propios datos: som.fit(X); luego usa u_matrix(), map_vectors(), etc.")
```
