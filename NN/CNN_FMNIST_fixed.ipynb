```python
# ==============================================================
# CNN + Fashion-MNIST (70k imagenes en escala de grises)
# Enfocada en: Convoluciones, Pooling y Fully-Connected
# - Hiperparámetros en una sola sección (CONFIG)
# - Arquitectura construida dinámicamente a partir de CONFIG
# - Curvas, matriz de confusión, filtros y activaciones
# ==============================================================

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import confusion_matrix, classification_report
import itertools, time

# ----------------------------
# 0) HIPERPARÁMETROS (EDITA AQUÍ)
# ----------------------------
CONFIG = {
    # Datos
    "dataset": "fashion_mnist",   # "fashion_mnist" (28x28x1)
    "val_size": 5000,

    # Entrenamiento
    "batch_size": 128,
    "epochs": 6,
    "optimizer": "adam",          # "adam", "sgd"
    "learning_rate": 1e-3,
    "weight_decay": 0.0,          # útil con AdamW (si quisieras)
    "label_smoothing": 0.0,

    # Augmentación (opcional para ver su efecto)
    "use_augmentation": True,
    "aug_params": {
        "random_flip": "horizontal",   # "horizontal" | "vertical" | None
        "random_rotation": 0.08,       # en radianes aprox (tf.image)
        "random_zoom": 0.08,           # fracción
        "random_translation": 0.05     # fracción
    },

    # Arquitectura: BLOQUES CONVOLUCIONALES
    # Cada bloque: Conv->BN->ReLU->(Pooling)
    "conv_blocks": [
        {"filters": 32, "kernel_size": 3, "padding": "same", "use_bias": False,
         "pool_type": "max", "pool_size": 2},  # "max" | "avg" | None
        {"filters": 64, "kernel_size": 3, "padding": "same", "use_bias": False,
         "pool_type": "max", "pool_size": 2},
    ],

    # Capa de transición antes de las densas
    # "global_pool": "gap" usa GlobalAveragePooling2D
    # "global_pool": "gmp" usa GlobalMaxPooling2D
    # None => hace Flatten (más parámetros)
    "global_pool": "gap",   # "gap" | "gmp" | None

    # Fully Connected (cabeza clasificadora)
    "dropout": 0.25,
    "dense_units": [64],    # puedes poner [128, 64], etc.
    "num_classes": 10,
}

CLASS_NAMES = ['T-shirt/top','Trouser','Pullover','Dress','Coat',
               'Sandal','Shirt','Sneaker','Bag','Ankle boot']

# ----------------------------
# 1) Datos
# ----------------------------
def load_data(cfg):
    if cfg["dataset"] == "fashion_mnist":
        (x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
        x_train = x_train.astype("float32")/255.0
        x_test  = x_test.astype("float32")/255.0
        # añadir canal
        x_train = x_train[..., np.newaxis]
        x_test  = x_test[..., np.newaxis]
    else:
        raise ValueError("Dataset no soportado en este ejemplo.")

    # split validación
    val_size = cfg["val_size"]
    x_val, y_val = x_train[-val_size:], y_train[-val_size:]
    x_train, y_train = x_train[:-val_size], y_train[:-val_size]
    return (x_train, y_train), (x_val, y_val), (x_test, y_test)

def build_augmenter(cfg):
    if not cfg["use_augmentation"]:
        return keras.Sequential([], name="no_aug")
    p = cfg["aug_params"]
    layers_list = []
    if p.get("random_flip"):
        layers_list.append(layers.RandomFlip(p["random_flip"]))
    if p.get("random_rotation", 0) > 0:
        layers_list.append(layers.RandomRotation(p["random_rotation"]))
    if p.get("random_zoom", 0) > 0:
        layers_list.append(layers.RandomZoom(p["random_zoom"]))
    if p.get("random_translation", 0) > 0:
        frac = p["random_translation"]
        layers_list.append(layers.RandomTranslation(frac, frac))
    return keras.Sequential(layers_list, name="augmenter")

# ----------------------------
# 2) Modelo CNN (Convolución, Pooling, Fully-Connected)
# ----------------------------
def build_model(cfg, input_shape):
    inputs = keras.Input(shape=input_shape)
    x = inputs

    # Augmentación en el grafo (opcional)
    aug = build_augmenter(cfg)
    x = aug(x)

    # BLOQUES CONV: Conv -> BN -> ReLU -> Pool (opcional)
    for i, b in enumerate(cfg["conv_blocks"], start=1):
        x = layers.Conv2D(
            filters=b["filters"],
            kernel_size=b["kernel_size"],
            padding=b["padding"],
            use_bias=b["use_bias"],
            name=f"conv{i}"
        )(x)
        x = layers.BatchNormalization(name=f"bn{i}")(x)
        x = layers.ReLU(name=f"relu{i}")(x)

        if b["pool_type"] == "max":
            x = layers.MaxPooling2D(pool_size=b["pool_size"], name=f"maxpool{i}")(x)
        elif b["pool_type"] == "avg":
            x = layers.AveragePooling2D(pool_size=b["pool_size"], name=f"avgpool{i}")(x)
        # si None, no se aplica pooling en este bloque

    # TRANSICIÓN HACIA DENSAS
    if cfg["global_pool"] == "gap":
        x = layers.GlobalAveragePooling2D(name="gap")(x)  # reduce HxW agregando por canal
    elif cfg["global_pool"] == "gmp":
        x = layers.GlobalMaxPooling2D(name="gmp")(x)
    else:
        x = layers.Flatten(name="flatten")(x)  # mantiene más info, más parámetros

    # FULLY CONNECTED (cabeza)
    if cfg["dropout"] and cfg["dropout"] > 0:
        x = layers.Dropout(cfg["dropout"], name="dropout")(x)
    for j, units in enumerate(cfg["dense_units"], start=1):
        x = layers.Dense(units, activation="relu", name=f"fc{j}")(x)

    outputs = layers.Dense(cfg["num_classes"], activation="softmax", name="logits")(x)
    model = keras.Model(inputs, outputs, name="cnn_modular")

    # Compilación
    if cfg["optimizer"].lower() == "adam":
        opt = keras.optimizers.Adam(learning_rate=cfg["learning_rate"])
    elif cfg["optimizer"].lower() == "sgd":
        opt = keras.optimizers.SGD(learning_rate=cfg["learning_rate"], momentum=0.9)
    else:
        raise ValueError("Optimizer no soportado")

    model.compile(
        optimizer=opt,
        loss=keras.losses.SparseCategoricalCrossentropy(), # Removed label_smoothing
        metrics=["accuracy"]
    )
    return model

# ----------------------------
# 3) Utilidades de visualización
# ----------------------------
def plot_examples(x, y, class_names, n=12):
    plt.figure(figsize=(10,3))
    for i in range(n):
        plt.subplot(2, n//2, i+1)
        plt.imshow(x[i].squeeze(), cmap='gray')
        plt.title(class_names[y[i]])
        plt.axis('off')
    plt.suptitle("Muestras del dataset")
    plt.show()

def plot_history(h):
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1); plt.plot(h.history['loss'], label='train'); plt.plot(h.history['val_loss'], label='val'); plt.title('Pérdida'); plt.legend()
    plt.subplot(1,2,2); plt.plot(h.history['accuracy'], label='train'); plt.plot(h.history['val_accuracy'], label='val'); plt.title('Exactitud'); plt.legend()
    plt.tight_layout(); plt.show()

def plot_confusion(cm, classes, normalize=True, title='Matriz de confusión'):
    if normalize:
        cm = cm.astype("float")/cm.sum(axis=1, keepdims=True)
    plt.figure(figsize=(6,6))
    plt.imshow(cm, interpolation='nearest')
    plt.title(title); plt.colorbar()
    ticks = np.arange(len(classes))
    plt.xticks(ticks, classes, rotation=45, ha='right'); plt.yticks(ticks, classes)
    fmt = '.2f' if normalize else 'd'; thresh = cm.max()/2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 ha="center", color="white" if cm[i, j] > thresh else "black", fontsize=8)
    plt.ylabel('Real'); plt.xlabel('Predicho'); plt.tight_layout(); plt.show()

def visualize_first_conv_filters(model):
    conv = [l for l in model.layers if isinstance(l, layers.Conv2D)][0]
    W = conv.get_weights()[0]  # (k,k,Cin,Cout)
    wmin, wmax = W.min(), W.max()
    K = (W - wmin)/(wmax - wmin + 1e-8)
    n_filters = K.shape[-1]; cols = 8; rows = int(np.ceil(n_filters/cols))
    plt.figure(figsize=(cols, rows))
    for i in range(n_filters):
        plt.subplot(rows, cols, i+1)
        plt.imshow(K[:,:,0,i], cmap='gray'); plt.axis('off')
    plt.suptitle("Filtros aprendidos (1ª conv)"); plt.show()

def visualize_activations(model, x_batch, layer_type=layers.Conv2D, img_idx=0):
    sample = x_batch[img_idx:img_idx+1]
    conv_layers = [l for l in model.layers if isinstance(l, layer_type)]
    if not conv_layers:
        print("No se encontraron capas del tipo solicitado."); return
    act_model = keras.Model(model.input, [l.output for l in conv_layers])
    activations = act_model.predict(sample, verbose=0)
    for idx, act in enumerate(activations, start=1):
        n_feat = act.shape[-1]; cols = 8; rows = int(np.ceil(n_feat/cols))
        plt.figure(figsize=(cols, rows))
        for i in range(n_feat):
            plt.subplot(rows, cols, i+1)
            plt.imshow(act[0,:,:,i], cmap='gray'); plt.axis('off')
        plt.suptitle(f"Mapas de activación capa conv #{idx}"); plt.show()

# ----------------------------
# 4) Run
# ----------------------------
(x_train, y_train), (x_val, y_val), (x_test, y_test) = load_data(CONFIG)
print("Train:", x_train.shape, y_train.shape, "| Val:", x_val.shape, y_val.shape, "| Test:", x_test.shape, y_test.shape)
plot_examples(x_train, y_train, CLASS_NAMES, n=12)

model = build_model(CONFIG, input_shape=x_train.shape[1:])
model.summary()

start = time.time()
history = model.fit(
    x_train, y_train,
    validation_data=(x_val, y_val),
    batch_size=CONFIG["batch_size"],
    epochs=CONFIG["epochs"],
    verbose=1
)
print(f"Entrenamiento en {time.time()-start:.1f} s")

plot_history(history)

test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print(f"Test accuracy: {test_acc:.4f}")

y_pred = np.argmax(model.predict(x_test, verbose=0), axis=1)
cm = confusion_matrix(y_test, y_pred)
plot_confusion(cm, CLASS_NAMES, normalize=True)

print(classification_report(y_test, y_pred, target_names=CLASS_NAMES, digits=4))

visualize_first_conv_filters(model)
visualize_activations(model, x_test, img_idx=3)
```

```python
# ==============================================================
# CNN + Fashion-MNIST (70k imágenes, 28x28x1)
# Versión mejorada: +filtros, kernel 5x5 inicial, 3er bloque conv,
# augmentación más fuerte. Colab-ready.
# ==============================================================

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import confusion_matrix, classification_report
import itertools, time

# ----------------------------
# 0) HIPERPARÁMETROS (EDITA AQUÍ)
# ----------------------------
CONFIG = {
    # Datos
    "dataset": "fashion_mnist",
    "val_size": 5000,

    # Entrenamiento
    "batch_size": 128,
    "epochs": 10,                 # subimos épocas
    "optimizer": "adam",
    "learning_rate": 1e-3,
    "weight_decay": 0.0,
    "label_smoothing": 0.0,

    # Augmentación (ligeramente más fuerte)
    "use_augmentation": True,
    "aug_params": {
        "random_flip": "horizontal",
        "random_rotation": 0.12,   # antes 0.08
        "random_zoom": 0.12,       # antes 0.08
        "random_translation": 0.08 # antes 0.05
    },

    # Arquitectura: BLOQUES CONVOLUCIONALES
    # Conv -> BN -> ReLU -> (Pooling)
    "conv_blocks": [
        {"filters": 48, "kernel_size": 5, "padding": "same", "use_bias": False,
         "pool_type": "max", "pool_size": 2},
        {"filters": 96, "kernel_size": 3, "padding": "same", "use_bias": False,
         "pool_type": "max", "pool_size": 2},
        {"filters": 128, "kernel_size": 3, "padding": "same", "use_bias": False,
         "pool_type": "avg", "pool_size": 2},   # 3er bloque + avg pooling
    ],

    # Transición antes de densas
    "global_pool": "gap",   # GAP mantiene pocos parámetros y buena generalización

    # Fully Connected
    "dropout": 0.30,        # un poco más de regularización
    "dense_units": [128, 64],
    "num_classes": 10,
}

CLASS_NAMES = ['T-shirt/top','Trouser','Pullover','Dress','Coat',
               'Sandal','Shirt','Sneaker','Bag','Ankle boot']

# ----------------------------
# 1) Datos
# ----------------------------
def load_data(cfg):
    if cfg["dataset"] == "fashion_mnist":
        (x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
        x_train = x_train.astype("float32")/255.0
        x_test  = x_test.astype("float32")/255.0
        x_train = x_train[..., np.newaxis]
        x_test  = x_test[..., np.newaxis]
    else:
        raise ValueError("Dataset no soportado en este ejemplo.")
    val_size = cfg["val_size"]
    x_val, y_val = x_train[-val_size:], y_train[-val_size:]
    x_train, y_train = x_train[:-val_size], y_train[:-val_size]
    return (x_train, y_train), (x_val, y_val), (x_test, y_test)

def build_augmenter(cfg):
    if not cfg["use_augmentation"]:
        return keras.Sequential([], name="no_aug")
    p = cfg["aug_params"]
    layers_list = []
    if p.get("random_flip"):
        layers_list.append(layers.RandomFlip(p["random_flip"]))
    if p.get("random_rotation", 0) > 0:
        layers_list.append(layers.RandomRotation(p["random_rotation"]))
    if p.get("random_zoom", 0) > 0:
        layers_list.append(layers.RandomZoom(p["random_zoom"]))
    if p.get("random_translation", 0) > 0:
        frac = p["random_translation"]
        layers_list.append(layers.RandomTranslation(frac, frac))
    return keras.Sequential(layers_list, name="augmenter")

# ----------------------------
# 2) Modelo CNN (Convolución, Pooling, Fully-Connected)
# ----------------------------
def build_model(cfg, input_shape):
    inputs = keras.Input(shape=input_shape)
    x = inputs

    # Augmentación en el grafo (opcional)
    x = build_augmenter(cfg)(x)

    # BLOQUES CONV: Conv -> BN -> ReLU -> Pool
    for i, b in enumerate(cfg["conv_blocks"], start=1):
        x = layers.Conv2D(
            filters=b["filters"],
            kernel_size=b["kernel_size"],
            padding=b["padding"],
            use_bias=b["use_bias"],
            name=f"conv{i}"
        )(x)
        x = layers.BatchNormalization(name=f"bn{i}")(x)
        x = layers.ReLU(name=f"relu{i}")(x)

        if b["pool_type"] == "max":
            x = layers.MaxPooling2D(pool_size=b["pool_size"], name=f"maxpool{i}")(x)
        elif b["pool_type"] == "avg":
            x = layers.AveragePooling2D(pool_size=b["pool_size"], name=f"avgpool{i}")(x)

    # TRANSICIÓN HACIA DENSAS
    if cfg["global_pool"] == "gap":
        x = layers.GlobalAveragePooling2D(name="gap")(x)
    elif cfg["global_pool"] == "gmp":
        x = layers.GlobalMaxPooling2D(name="gmp")(x)
    else:
        x = layers.Flatten(name="flatten")(x)

    # FULLY CONNECTED
    if cfg["dropout"] and cfg["dropout"] > 0:
        x = layers.Dropout(cfg["dropout"], name="dropout")(x)
    for j, units in enumerate(cfg["dense_units"], start=1):
        x = layers.Dense(units, activation="relu", name=f"fc{j}")(x)

    outputs = layers.Dense(cfg["num_classes"], activation="softmax", name="logits")(x)
    model = keras.Model(inputs, outputs, name="cnn_modular")

    # Optimizador
    if cfg["optimizer"].lower() == "adam":
        opt = keras.optimizers.Adam(learning_rate=cfg["learning_rate"])
    elif cfg["optimizer"].lower() == "sgd":
        opt = keras.optimizers.SGD(learning_rate=cfg["learning_rate"], momentum=0.9)
    else:
        raise ValueError("Optimizer no soportado")

    model.compile(
        optimizer=opt,
        loss=keras.losses.SparseCategoricalCrossentropy(),  # sin label smoothing aquí
        metrics=["accuracy"]
    )
    return model

# ----------------------------
# 3) Utilidades de visualización
# ----------------------------
def plot_examples(x, y, class_names, n=12):
    plt.figure(figsize=(10,3))
    for i in range(n):
        plt.subplot(2, n//2, i+1)
        plt.imshow(x[i].squeeze(), cmap='gray')
        plt.title(class_names[y[i]])
        plt.axis('off')
    plt.suptitle("Muestras del dataset")
    plt.show()

def plot_history(h):
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1); plt.plot(h.history['loss'], label='train'); plt.plot(h.history['val_loss'], label='val'); plt.title('Pérdida'); plt.legend()
    plt.subplot(1,2,2); plt.plot(h.history['accuracy'], label='train'); plt.plot(h.history['val_accuracy'], label='val'); plt.title('Exactitud'); plt.legend()
    plt.tight_layout(); plt.show()

def plot_confusion(cm, classes, normalize=True, title='Matriz de confusión'):
    if normalize:
        cm = cm.astype("float")/cm.sum(axis=1, keepdims=True)
    plt.figure(figsize=(6,6))
    plt.imshow(cm, interpolation='nearest')
    plt.title(title); plt.colorbar()
    ticks = np.arange(len(classes))
    plt.xticks(ticks, classes, rotation=45, ha='right'); plt.yticks(ticks, classes)
    fmt = '.2f' if normalize else 'd'; thresh = cm.max()/2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 ha="center", color="white" if cm[i, j] > thresh else "black", fontsize=8)
    plt.ylabel('Real'); plt.xlabel('Predicho'); plt.tight_layout(); plt.show()

def visualize_first_conv_filters(model):
    conv = [l for l in model.layers if isinstance(l, layers.Conv2D)][0]
    W = conv.get_weights()[0]  # (k,k,Cin,Cout)
    wmin, wmax = W.min(), W.max()
    K = (W - wmin)/(wmax - wmin + 1e-8)
    n_filters = K.shape[-1]; cols = 8; rows = int(np.ceil(n_filters/cols))
    plt.figure(figsize=(cols, rows))
    for i in range(n_filters):
        plt.subplot(rows, cols, i+1)
        plt.imshow(K[:,:,0,i], cmap='gray'); plt.axis('off')
    plt.suptitle("Filtros aprendidos (1ª conv)"); plt.show()

def visualize_activations(model, x_batch, layer_type=layers.Conv2D, img_idx=0):
    sample = x_batch[img_idx:img_idx+1]
    conv_layers = [l for l in model.layers if isinstance(l, layer_type)]
    if not conv_layers:
        print("No se encontraron capas del tipo solicitado."); return
    act_model = keras.Model(model.input, [l.output for l in conv_layers])
    activations = act_model.predict(sample, verbose=0)
    for idx, act in enumerate(activations, start=1):
        n_feat = act.shape[-1]; cols = 8; rows = int(np.ceil(n_feat/cols))
        plt.figure(figsize=(cols, rows))
        for i in range(n_feat):
            plt.subplot(rows, cols, i+1)
            plt.imshow(act[0,:,:,i], cmap='gray'); plt.axis('off')
        plt.suptitle(f"Mapas de activación capa conv #{idx}"); plt.show()

# ----------------------------
# 4) Run
# ----------------------------
(x_train, y_train), (x_val, y_val), (x_test, y_test) = load_data(CONFIG)
print("Train:", x_train.shape, y_train.shape, "| Val:", x_val.shape, y_val.shape, "| Test:", x_test.shape, y_test.shape)
plot_examples(x_train, y_train, CLASS_NAMES, n=12)

model = build_model(CONFIG, input_shape=x_train.shape[1:])
model.summary()

# (Opcional) callback para bajar LR si se estanca
callbacks = [keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=2, verbose=1)]

start = time.time()
history = model.fit(
    x_train, y_train,
    validation_data=(x_val, y_val),
    batch_size=CONFIG["batch_size"],
    epochs=CONFIG["epochs"],
    callbacks=callbacks,
    verbose=1
)
print(f"Entrenamiento en {time.time()-start:.1f} s")

plot_history(history)

test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print(f"Test accuracy: {test_acc:.4f}")

y_pred = np.argmax(model.predict(x_test, verbose=0), axis=1)
cm = confusion_matrix(y_test, y_pred)
plot_confusion(cm, CLASS_NAMES, normalize=True)

print(classification_report(y_test, y_pred, target_names=CLASS_NAMES, digits=4))

visualize_first_conv_filters(model)
visualize_activations(model, x_test, img_idx=3)
```
